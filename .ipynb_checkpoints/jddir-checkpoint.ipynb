### 将id放到最后，缀入risk列，dummy scan, drop sec、log_id、timelong、time

```python
scanDummy = pd.get_dummies(t_login['is_scan'],prefix='scan')
t_login_mix = t_login.drop(['log_id','timelong','is_scan','time','is_sec'],1)
t_login_mix = pd.concat([t_login_mix,scanDummy],axis=1)
id = t_login_mix.pop('id')
t_login_mix.insert(9,'id',id)
t_login_mix.insert(10,'risk',0)
print t_login_mix.columns
t_login_mix_np = t_login_mix.as_matrix()
x = t_login_mix_np[:,0:-1]
y = t_login_mix_np[:,-1]
pass
```
